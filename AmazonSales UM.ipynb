```python
# Transforming 'Order Date' & 'Ship Date' column to datetime format 
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data.dtypes
```

```python
# Convert 'Order Date' & 'Ship Date' column to datetime format 
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data.dtypes
```

```python
# Extracting month and year from 'Order Date' column
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
```

```python
# Calculate the number of 'Regions','Country'Item Types
region = data['Region'].nunique()
country = data['Country'].nunique()
item_type = data['Item Type'].nunique()
units_sold = data['Units Sold'].sum()
unit_cost = data['Unit Cost'].sum()
total_revenue = data['Total Revenue'].sum()
total_cost = data['Total Cost'].sum()
total_profit = data['Total Profit'].sum()
print('No. of Regions',region)
print('No. of Country',country)
print('No. of Item Types',item_type)
print('No. of Unit Sold',units_sold)
print('No. of Units Cost',unit_cost)
print('No. of Total Revenue',total_revenue)
print('No. of Total Cost',total_cost)
print('No. of Total Profit',total_profit)
```

```python
data.groupby(['Region','Sales Channel'])['Total Profit'].sum()
```

```python
# Grouping data by month and year and yearly monthly calculating total sales
monthly_sales = data.groupby(['Order Year', 'Order Month'])['Total Revenue'].sum().reset_index()
yearly_sales = data.groupby('Order Year')['Total Revenue'].sum().reset_index()
yearly_month_sales = data.groupby(['Order Year', 'Order Month'])['Total Revenue'].mean().unstack()
```

```python
# Year wise Sales Trend
plt.figure(figsize=(10, 6))
sns.barplot(x='Order Year', y='Total Revenue', data=yearly_sales)
plt.title('Year-wise Sales Trends')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.show()
```

According to this graph 2012 has high Sales and 2011 has low Sales

```python
# Plotting yearly_monthly sales trend
yearly_monthly_sales = data.groupby(['Order Year', 'Order Month'])['Total Revenue'].sum().reset_index()

plt.figure(figsize=(12, 8))
sns.lineplot(data=yearly_monthly_sales, x='Order Month', y='Total Revenue', hue='Order Year', marker='o')
plt.title('Yearly Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year')
plt.grid(True)
plt.show()
```

In july 2012 they have High Sales and
In June 2011, August 2017 and September 2014 They have low Sales

```python
# Groupby Total Profit by Region
TotalProfit_Region = data.groupby('Region')['Total Profit'].mean()
```

```python
# Pie chart of Total Profit by Region
plt.figure(figsize=(8,8))
TotalProfit_Region.plot(kind='pie',autopct='%1.1f%%',startangle=90)
plt.title('Total Profit by Region')
```

Region wise Middle East And North Africa has high profit

```python
# Groupby Total Revenue by Item Type
TotalRevenue_ItemType = data.groupby('Item Type')['Total Revenue'].sum()
```

```python
# Bar chart for Total Revenue by Item Type
plt.figure(figsize=(8,3))
TotalRevenue_ItemType.plot(kind='bar')
plt.xlabel('Item Type')
plt.ylabel('Total Revenue')
plt.title('Average Revenue by Product type')
plt.grid(axis='y')
```

Cosmetics has high revenue

```python
# Groupby Total Revenue by Item Type
TotalRevenue_SalesChannel = data.groupby('Sales Channel')['Total Revenue'].mean()
```

```python
# Pie chart for Total Revenue by Item Type
plt.figure(figsize=(6,6))
plt.tight_layout()
TotalRevenue_SalesChannel.plot(kind='pie',autopct='%1.1f%%',startangle=90)
plt.title('Total Revenue by Sales Channel')
```

Offline has high revenue

```python
#Key metrics and factor analysis

key_metrics = data[['Total Revenue', 'Total Cost', 'Total Profit', 'Units Sold']]

# Descriptive statistics
print(key_metrics.describe())

# Correlation matrix
correlation_matrix = key_metrics.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Key Metrics')
plt.show()
```

```python
# Visualizing sales by other factors as needed
# Example: Sales by Order Priority
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Order Priority', y='Total Revenue', estimator=sum, ci=None)
plt.title('Total Sales by Order Priority')
plt.xlabel('Order Priority')
plt.ylabel('Total Sales')
plt.show()
```
